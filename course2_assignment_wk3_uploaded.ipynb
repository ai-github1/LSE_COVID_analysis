```python
# "calculate differences between the application of the 
# first and second dose"

# They are daily (non-cumulative) time series so it seems
# to me that any day to day comparison is meaningless
# therefore I will look at totals for first and second dose

# create new df just for doses
doses_country = cases_vac[['Province/State',
 'Date',
 'First Dose', 'Second Dose']]
doses_country
```

```python
# apply aggregation for sum

total_doses_country = doses_country.groupby('Province/State')[['First Dose', 'Second Dose']].sum()
total_doses_country
```

```python
# add new column to highlight differences
total_doses_country['No Second Dose'] = total_doses_country['First Dose'] - total_doses_country['Second Dose']
total_doses_country
```

```python
# add new column to express as percentage of first dosed
total_doses_country['No Second Dose pct'] = 100*(total_doses_country['No Second Dose'] / total_doses_country['First Dose'])
total_doses_country
```

```python
# find max and min values in absolute and relative terms
total_doses_maxmin = total_doses_country.agg(['min', 'max'])
total_doses_maxmin
```

```python
total_doses_country.max(axis=0)
```

```python
print("Highest cases of No Second Dose")
print(total_doses_country['No Second Dose'].idxmax())
print(total_doses_country['No Second Dose'].max())
print(" ")
print("Lowest cases of No Second Dose")
print(total_doses_country['No Second Dose'].idxmin())
print(total_doses_country['No Second Dose'].min())
```

```python
print("Highest percent of No Second Dose")
print(total_doses_country['No Second Dose pct'].idxmax())
print(total_doses_country['No Second Dose pct'].max())
print(" ")
print("Lowest percent of No Second Dose")
print(total_doses_country['No Second Dose pct'].idxmin())
print(total_doses_country['No Second Dose pct'].min())
print(" ")
print("NB percent of No Second Dose = (no second dose / first dose)")
```

```python
# investigating total deaths and hospitalisations

# new subset of data
hospitaldata_country = cases_vac[['Province/State',
 'Date',
 'Deaths', 'Hospitalised', 'Recovered']]

# aggregate the data
total_hospitaldata_country = hospitaldata_country.groupby('Province/State')[['Deaths', 'Hospitalised', 'Recovered']].sum()

# print the max/min 
print("Highest deaths")
print(total_hospitaldata_country['Deaths'].idxmax())
print(total_hospitaldata_country['Deaths'].max())
print(" ")
print("Lowest deaths")
print(total_hospitaldata_country['Deaths'].idxmin())
print(total_hospitaldata_country['Deaths'].min())
print(" ")
print("Highest hospitalised")
print(total_hospitaldata_country['Hospitalised'].idxmax())
print(total_hospitaldata_country['Hospitalised'].max())
print(" ")
print("Lowest hospitalised")
print(total_hospitaldata_country['Hospitalised'].idxmin())
print(total_hospitaldata_country['Hospitalised'].min())
print(" ")
print("Highest Recovered")
print(total_hospitaldata_country['Recovered'].idxmax())
print(total_hospitaldata_country['Recovered'].max())
print(" ")
print("Lowest Recovered")
print(total_hospitaldata_country['Recovered'].idxmin())
print(total_hospitaldata_country['Recovered'].min())
```
